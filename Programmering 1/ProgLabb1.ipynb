```python
#UPPGIFT 1e

import random
import matplotlib.pyplot as plt
import math

plt.xlabel("x")
plt.ylabel("y")

cirkel = 0

for s in range (501):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)

    avstånd = math.sqrt(x**2 + y**2)

     
    if avstånd < 1:
        cirkel += 1
        plt.plot(x,y,'*r')
    else:
        plt.plot(x,y,'*b')

print(f"{cirkel}/{s}")
```

```python
#UPPGIFT2e

start = int(input("Ange start nummer"))
slut = int(input("Ange slut nummer"))
tal = int(input("Ange ditt Burr nummer"))
ta = int(input("Ange ditt birr nummer"))

burr = 0
birr = 0
burrbirr = 0

i = 0
for i in range(start,slut):
    if i%ta==0 and i%tal==0:
        print("burrbirr")
        burrbirr += 1
    elif i%tal==0:
        print("burr")
        burr += 1
    elif i%ta==0:
        print("birr")
        birr += 1
    else:
        print(i)

print(f"Antalet Burr är {burr}")
print(f"Antalet Birr är {birr}")
print(f"Antalet BurrBirr är {burrbirr}")
    
```
